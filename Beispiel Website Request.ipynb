```python
# importiert Requests
# installiert die Library BeautifulSoup – saugt Daten aus html!
# installiert Pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# mit json Zahl ausgeben
r = requests.get("https://www.srf.ch")
r = r.text
```

```python
soup = BeautifulSoup(r, 'lxml')
```

```python
len(soup.find_all ('h3'))
```

```python
lst = soup.find_all ('h3')[13:][:50]
```

```python
len(lst)
```

```python
fertigeliste = []
for artikel in lst:
    art = "".join(artikel.text.strip().split())
    fertigeliste.append(art)
```

```python
df = pd.DataFrame(fertigeliste)
```

```python
df.to_csv('srf_titelliste.csv')
```

```python
# Lösung von Lea
# Attribute, wie span, class oder title, usw, immer als string in Anführungszeichen angeben

for titel in soup.find_all('span', {'class':"title"}):
    print(titel.text)
```

```python
# listet alles h3 über 13 und unter 50

lst = soup.find_all ('h3')[13:][:50]
```
