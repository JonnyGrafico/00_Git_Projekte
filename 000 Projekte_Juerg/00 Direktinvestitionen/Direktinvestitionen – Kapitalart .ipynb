# DI – Beteiligungskapital OHNE reinvestierte Erträge

```python
df = pd.read_excel('000_snb-di_ohne_reinvestierte.xlsx', index_col=0, header=1)
# df = pd.read_excel('000_snb-di_ohne_reinvestierte.xlsx')
```

```python
df.fillna(0).head()
```

```python
df.columns
```

```python
df_ohnere_total = df[['Industrie - Total', 'Dienste - Total', 'Total',
       'Total ohne ausländisch beherrschte Finanz- und Holdinggesellschaften',
       #'Pro memoria: Grundstücks- und Wohnungswesen',
       #'Pro memoria: Total inkl. Grundstücks- und Wohnungswesen'
        ]]
df_ohnere_total.iloc[8:19].apply(lambda x: x/1000).plot(
    kind='bar', figsize=(15,10),
    fontsize=14,
    title="DI: Beteiligungskapital ohne reinvestierte Erträge – Total")
```

```python
df_ohnere_dienste = df[['Dienste - Handel',
       'Dienste - Finanz- und Holdinggesellschaften - Total',
       'Dienste - Finanz- und Holdinggesellschaften - schweizerisch beherrscht',
       'Dienste - Finanz- und Holdinggesellschaften - ausländisch beherrscht - Total',
       'Dienste - Finanz- und Holdinggesellschaften - ausländisch beherrscht - davon Special Purpose Entities',
       'Dienste - Banken', 'Dienste - Versicherungen',
       'Dienste - Transporte und Kommunikation', 'Dienste - Übrige Dienste']]
df_ohnere_dienste.iloc[8:19].apply(lambda x: x/1000).plot(
    kind='bar', figsize=(15,10),
    fontsize=14,
    title="DI: Beteiligungskapital ohne reinvestierte Erträge – DIENSTE")
```

```python
df_ohnere_industrie = df[['Industrie - Textilien und Bekleidung',
       'Industrie - Chemie und Kunststoffe',
       'Industrie - Metalle und Maschinen',
       'Industrie - Elektronik, Energie, Optik und Uhren',
       'Industrie - Übrige Industrien und Bau']]
df_ohnere_industrie.iloc[8:19].apply(lambda x: x/1000).plot(
    kind='bar', figsize=(15,10),
    fontsize=14,
    title="DI: Beteiligungskapital ohne reinvestierte Erträge – INDUSTRIE")
```
