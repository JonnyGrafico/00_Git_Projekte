## Europa

```python
#df = pd.read_excel('snb-data-fdiaustabsa-de-all-20171215_0900.xlsx', index_col=0, header=0)
df = pd.read_excel('snb-data-fdiaustlanda-de-all-20171215_0900.xlsx')
```

```python
df = df.set_index('Jahre')
df.fillna(0).head(5)
```

```python
#df.columns
```

```python
df.columns = df.columns.str.strip().str.replace(
    'Alle Länder - Europa - ', '_').str.replace(
    'Alle Länder - Nordamerika - ', '_').str.replace(
    'Alle Länder - Mittel- und Südamerika - ', '_').str.replace(
    'Alle Länder - Asien - ', '_').str.replace(
    'Alle Länder - Afrika - ', '_').str.replace(
    'Alle Länder - Ozeanien - ', '_').str.replace(
    '(', '').str.replace('', '').str.replace("'", "")

df.columns = df.columns.str.strip('_').str.strip(')')
```

```python
#df.columns
```

```python
europa = df[['Baltische Länder', 'Belgien',
       'Bulgarien', 'Dänemark', 'Deutschland', 'Finnland', 'Frankreich',
       'Griechenland', 'Irland', 'Italien', 'Kroatien', 'Luxemburg',
       'Niederlande', 'Norwegen', 'Österreich', 'Polen', 'Portugal',
       'Rumänien', 'Russische Föderation', 'Schweden', 'Slowakei', 'Spanien',
       'Tschechische Republik', 'Türkei', 'Ukraine', 'Ungarn',
       'Vereinigtes Königreich', 'Offshore-Finanzzentren Europa', 'EU',
       'Übriges Europa']]

europa.iloc[15:19].apply(lambda x: x/1000).plot(kind='bar', figsize=(10,8))
```

```python
nordamerika = df[['Kanada', 'Vereinigte Staaten']]
nordamerika.iloc[0:19].apply(lambda x: x/1000).plot(kind='bar', figsize=(10,8))
```

```python
mittel_und_südamerika = df[['Argentinien', 'Bolivien', 'Brasilien', 'Chile', 'Costa Rica',
       'Ecuador', 'Guatemala', 'Kolumbien', 'Mexiko', 'Peru', 'Uruguay',
       'Venezuela', 'Offshore-Finanzzentren Mittel- und Südamerika']]

mittel_und_südamerika.iloc[15:19].apply(lambda x: x/1000).plot(kind='bar', figsize=(10,8))
```

```python
asien = df[['Bangladesch', 'China', 'Hongkong', 'Indien', 'Indonesien', 'Israel',
       'Japan', 'Korea Republik Süd-', 'Malaysia', 'Pakistan', 'Philippinen',
       'Saudi-Arabien', 'Singapur', 'Sri Lanka', 'Taiwan', 'Thailand',
       'Vereinigte Arabische Emirate', 'Vietnam']]
asien.iloc[15:19].apply(lambda x: x/1000).plot(kind='bar', figsize=(10,8))
```
