```python
#df = pd.read_excel('snb-data-fdiaustabsa-de-all-20171215_0900.xlsx', index_col=0, header=0)
df = pd.read_excel('snb-data-fdiaustabsa-de-all-20171215_0900.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.set_index('Jahre')
df.fillna(0).head(5)
```

```python
df[['Industrie', 'Dienste']].head(5)
```

```python
df[['Industrie','Dienste']] = df[['Industrie','Dienste']].apply(lambda x: x/1000)
```

```python
df.head()
```

```python
# Draw a graph with pandas and keep what's returned
ax = df.plot(kind='bar', x='Jahre',
             y=['Industrie', 'Dienste'],
             figsize=(15,10), fontsize=14)

# Set the x scale because otherwise it goes into weird negative numbers
# ax.set_xlim((0, 70000))

# Set the x-axis label
ax.set_xlabel("Jahre", fontname="Arial", fontsize=14, fontweight='bold')

# Set the y-axis label
ax.set_ylabel("Milliarden Franken", fontsize=14)

# Set the y-axis titel
ax.set_title("Direktinvestitionen: Industrie und Dienste", fontname="Arial", fontsize=22)

ax.grid(b=None, which='major', axis='both', color='#DFEBEC', linestyle='--', linewidth=1)

ax.set_axisbelow(True)

```
