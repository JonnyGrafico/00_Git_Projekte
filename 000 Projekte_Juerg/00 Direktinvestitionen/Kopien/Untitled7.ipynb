```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = pd.read_excel('snb-data-fdiaustabsa-de-all-20171215_0900.xlsx', index_col=0, header=0)
df = pd.read_excel('snb-data-fdiaustabsa-de-all-20171215_0900.xlsx')
```

```python
#df.head()
```

```python
#df.tail()
```

```python
df = df.set_index('Jahre')
df.fillna(0).head(5)
```

```python
fig = df.iloc[0:19].apply(lambda x: x/1000).plot(
    kind='bar',
#     x='Jahre',
    y=['Total.3',
       'Total ohne ausländisch beherrschte Finanz- und Holdinggesellschaften.3'], 
    figsize=(25,15),
    fontsize=20,
    grid=True,
    title='Direktinvestitionen: Total',
    legend=True)
    
fig.set_title('Direktinvestitionen: Total', fontname="Verdana", fontsize=30, pad=30)
fig.legend(loc='center left', bbox_to_anchor=(0,-0.2), fontsize=20)
    
fig.set_ylabel("Milliarden Franken", fontsize=20)
fig.set_xlabel("Jahre", fontname="Arial", fontsize=20, fontweight='bold')
    
fig.grid(b=None, which='major', axis='both', color='#D4D9DB', linestyle='dotted', linewidth=2)
fig.set_axisbelow(True)
    
fig.set_facecolor('#F4F9FB')
```
